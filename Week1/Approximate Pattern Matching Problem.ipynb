```python
def Hamming_distance(Seq1,Seq2):
    Distance = 0
    for i in range(len(Seq1)):
        if(Seq1[i] != Seq2[i]):
            Distance += 1
    return Distance
```

```python
def ApproximatePatternMatching(Pattern,Sequence,d):
    k = len(Pattern)
    n = len(Sequence)
    Positions = set()
    for i in range(n-k+1):
        kmer = Sequence[i:i+k]
        Distance = Hamming_distance(kmer,Pattern)
        if(Distance<=d):
            Positions.add(i)
    return Positions
            
        
```

```python
Pattern = 'ATTCTGGA'
Sequence = 'CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAAT'
d = 3
ApproximatePatternMatching(Pattern,Sequence,d)
```

```python
Pattern = ''
Sequence = ''
d = 0
file = open('dataset_369239_4.txt', 'r') 
for i, line in enumerate(file):
    line=line.rstrip('\n')
    if(i==0):
        Pattern= line
    elif(i==1):
        Sequence = line
    else:
        d = int(line)
```
