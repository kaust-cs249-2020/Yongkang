```python
sequence = ''
file = open('dataset_369234_2.txt', 'r') 
for i, line in enumerate(file):
    line=line.rstrip('\n')
    sequence = line
```

```python
reverse_dict = {'A':'T','T':'A','C':'G','G':'C'}
```

```python
sequence_rc = ''
for base in sequence:
    sequence_rc = reverse_dict[base]+sequence_rc
    
```
