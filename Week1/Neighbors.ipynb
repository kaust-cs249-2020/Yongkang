```python
nucleotides = {'A','C','G','T'}
```

```python
def Hamming_distance(Seq1,Seq2):
    Distance = 0
    for i in range(len(Seq1)):
        if(Seq1[i] != Seq2[i]):
            Distance += 1
    return Distance
```

```python
def Neighbors(Pattern, d):
    if d == 0:
        return {Pattern}
    if len(Pattern) == 1: 
        return nucleotides
    Neighborhood = set()
    SuffixNeighbors = Neighbors(Pattern[1:len(Pattern)], d)
    for Text in SuffixNeighbors:
        if Hamming_distance(Pattern[1:len(Pattern)], Text) < d:
            for x in nucleotides:
                Neighborhood.add(x+Text)
        else:
            Neighborhood.add(Pattern[0]+Text)
    return Neighborhood
```

```python
Pattern = 'ACG'
d=1
```

```python
for pattern in Neighbors(Pattern, d):
    print(pattern,end=' ')
```

```python
Pattern = ''
d = 0
file = open('dataset_369242_4.txt', 'r') 
for i, line in enumerate(file):
    line=line.rstrip('\n')
    if(i==0):
        Pattern= line
    else:
        d = int(line)
```

```python
for pattern in Neighbors(Pattern, d):
    print(pattern,end=' ')
```
