```python
ComplRev = {'A':'T','T':'A','G':'C','C':'G'}
```

```python
def ReCom(kmer):
    rekmer = ''
    for base in kmer:
        rekmer = ComplRev[base]+rekmer
    return rekmer
```

```python
def shared_kmers(text1,text2,k):
    all = {}
    n = len(text1)
    for i in range(n-k+1):
        kmer = text1[i:i+k]
        if kmer in all.keys():
            all[kmer][0].add(i)
        elif ReCom(kmer) in all.keys():
            all[ReCom(kmer)][0].add(i)
        else:
            all[kmer] = [ {i},set() ]
    m = len(text2)
    for i in range(m-k+1):
        kmer = text2[i:i+k]
        if kmer in all.keys():
            all[kmer][1].add(i)
        elif ReCom(kmer) in all.keys():
            all[ReCom(kmer)][1].add(i)
        else:
            all[kmer] = [ set(),{i} ]
    pairs = []
    for item in all.values():
        if item[0] and item[1]:
            for i in item[0]:
                for j in item[1]:
                    pairs.append((i,j))
    return pairs
```

```python
k=3
text1 = 'AAACTCATC'
text2 = 'TTTCAAATC'
```

```python
shared_kmers(text1,text2,k)
```

```python
file = open('dataset_369339_5.txt', 'r') 
k=0
text1=''
text2=''
for i, line in enumerate(file):
    line=line.rstrip('\n')
    if(i==0):
        k = int(line)
    elif(i==1):
        text1 = line
    else:
        text2 = line
        
```
