```python
def ColoredEdges(P):
    Edges = []
    for idx,Chromosome in P.items():
        Nodes = ChromosomeToCycle(Chromosome)
        n = len(Chromosome)
        for j in range(n):
            if(j<n-1):
                Edges.append((int(Nodes[2*j+1]),int(Nodes[2*j+2])))
            else:
                Edges.append((int(Nodes[2*j+1]),int(Nodes[0])))
    return Edges
```

```python
def ChromosomeToCycle(Chromosome):
    n = len(Chromosome)
    Nodes = np.zeros(2*n)
    for j in range(n):
        i = Chromosome[j]
        if i > 0:
            Nodes[2*j] = int(2*i-1)
            Nodes[2*j+1]  =  int(2*i)
        else:
            Nodes[2*j]  = int(-2*i)
            Nodes[2*j+1]  =  int(-2*i-1)
    return Nodes
```
