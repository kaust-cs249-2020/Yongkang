```python
def  PathToGenome(kmers):
    Text = kmers[0]
    n = len(kmers)
    for i in range(1,n):
        Text += kmers[i][-1]
    return Text
```

```python
kmers = ['ACCGA',
'CCGAA',
'CGAAG',
'GAAGC',
'AAGCT']

```

```python
PathToGenome(kmers)
```

```python
kmers = []
file = open('dataset_369268_3.txt', 'r') 
for i, line in enumerate(file):
    line=line.rstrip('\n')
    kmers.append(line)
```

```python
Dna = PathToGenome(kmers)
```

```python
out = open('file','w')
out.write('%s\n'%(Dna))
```
