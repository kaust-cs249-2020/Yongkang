```python
def DeBruijnFromkMers(Patterns):
    dB = {}
    for pattern in Patterns:
        prefix = pattern[0:-1]
        suffix = pattern[1:]
        if prefix in dB.keys():
            dB[prefix] += ','+suffix
        else:
            dB[prefix] = suffix
    return dB
```

```python
Patterns = ['GAGG',
'CAGG',
'GGGG',
'GGGA',
'CAGG',
'AGGG',
'GGAG'
]
```

```python
dB =  DeBruijnFromkMers(Patterns)
for kmer1,kmer2 in dB.items():
    print("%s -> %s"%(kmer1,kmer2))
```

```python
Patterns = []
file = open('dataset_369270_8.txt', 'r') 
for i, line in enumerate(file):
    line=line.rstrip('\n')
    Patterns.append(line)
```

```python
dB =  DeBruijnFromkMers(Patterns)
out = open('file','w')
for kmer1,kmer2 in dB.items():
    out.write("%s -> %s\n"%(kmer1,kmer2))
```
