```python
def Find_minimum_element(DStart):
    mini = float("inf")
    i_min = j_min = None
    for i in DStart:
        for j in DStart[i]:
            if DStart[i][j] < mini:
                mini = DStart[i][j]
                i_min, j_min = i, j
    return i_min,j_min
```

```python
def reBuild(D,i_min,j_min):
    rows = list(D)
    m = max(D) + 1
    D[m] = {}
    D[m][m] = 0.0
    for k in rows:
        if k != i_min and k != j_min:
            D[m][k] = (D[i_min][k] + D[j_min][k] - D[i_min][j_min])/2.0
            D[k][m] = D[m][k]
            del D[k][i_min]
            del D[k][j_min]
    del D[i_min]
    del D[j_min]
    return D,m
```

```python
def neighborJoining(D, n):
    if n == 2:
        T  = dict()
        for i in D:
            for j in D[i]:
                if  j == i:
                    continue
                T[i] = [(j,D[i][j])]  
                T[j] = [(j,D[i][j])] 
        return T
        
    # Construct neighbor-joining D D* from D.
    totalDistance = {}
    for i in D:
        totalDistance[i] = 0
        for j in D[i]:
            totalDistance[i] += D[i][j]
    DStart = {}
    for i in D:
        DStart[i] = {}
        for j in D[i]:
            if i == j:
                DStart[i][j] = 0
            else:
                DStart[i][j] = (n-2)*D[i][j] - totalDistance[i] - totalDistance[j]
    
    
    #find elements i and j such that D*i,j is a minimum non-diagonal element of D*
    i_min,j_min = Find_minimum_element(DStart)
   
    #Δ ← (TotalDistanceD(i) - TotalDistanceD(j)) /(n - 2)
    delta = (totalDistance[i_min] - totalDistance[j_min])/(n-2.0)
    #limbLengthi ← (1/2)(Di,j + Δ)
    #limbLengthj ← (1/2)(Di,j - Δ)
    limbLengthi_min = (D[i_min][j_min] + delta)/2.0
    limbLengthj_min = (D[i_min][j_min] - delta)/2.0
    
   
    #add a new row/column m to D so that Dk,m = Dm,k = (1/2)(Dk,i + Dk,j - Di,j) for any k
    #D ← D with rows i and j removed
    #D ← D with columns i and j removed
    D,newNode= reBuild(D,i_min,j_min)       
    
    #T ← NeighborJoining(D)
    #edge, weight = neighborJoining(D, n-1)
    T = neighborJoining(D, n-1)
    # add two new limbs (connecting node m with leaves i and j) to the tree T
    #assign length limbLengthi to Limb(i)
    #assign length limbLengthj to Limb(j)
    if i_min in T.keys():
        T[i_min].append((newNode,limbLengthi_min))
    else:
        T[i_min] = [(newNode,limbLengthi_min)]
        
    if j_min in T.keys():
        T[j_min].append((newNode,limbLengthj_min))
    else:
        T[j_min] = [(newNode,limbLengthj_min)]
        
    if newNode in T.keys():
        T[newNode].append((i_min,limbLengthi_min))
        T[newNode].append((j_min,limbLengthj_min))
    else:
        T[newNode] = [(i_min,limbLengthi_min),(j_min,limbLengthj_min)]
    return T
```

```python
n = 0
distanceMatrix = {}
file = open('dataset_369353_7.txt', 'r') 
for i, line in enumerate(file):
    line=line.strip()
    if(i==0):
        n = int(line)
    else:
        distanceMatrix[i-1] = {}
        data = line.split(' ')
        for j in range(len(data)):
            distanceMatrix[i-1][j] = float(data[j])
```

```python
T = neighborJoining(distanceMatrix, n)
```

```python
for i in sorted (T.keys()):
    for leaf in T[i]:
        node,weight = leaf
        print("%d->%d:%.3f"%(i,node,weight))
```
