```python
import numpy as np
import queue
```

```python
n =4
AdjacencyDict = {0:[(4,11)],
                 1:[(4,2)],
                 2:[(5,6)],
                3:[(5,7)],
                4:[(0,11),(1,2),(5,4)],
                5:[(4,4),(3,7),(2,6)]}
```

```python
AdjacencyDict
```

```python
def calculatedistanceMatrix(n, AdjacencyDict):
    distanceMatrix = np.zeros([n,n]).astype(int)
    for i in range(n):
        dist = dict()
        q = queue.Queue()
        dist[i] = 0
        q.put(i)
        while not q.empty():
            currNode = q.get()
            for node, weight in AdjacencyDict[currNode]:
                if not node in dist:
                    dist[node] = dist[currNode] + weight
                    if node < n:
                        distanceMatrix[i][node] = dist[node]
                    q.put(node)
    return distanceMatrix
                        
```

```python
distanceMatrix = calculateDistMatrix(n, AdjacencyDict)
```

```python
for d in distanceMatrix:
    print(' '.join([str(i) for i in d]))
```

```python
n = 0
AdjacencyDict = dict()
file = open('dataset_369348_12.txt', 'r') 
for i, line in enumerate(file):
    line=line.rstrip('\n')
    if(i==0):
        n = int(line)
    else:
        edge = line.split('->')
        node1 = int(edge[0])
        node2,weight = edge[1].split(':')
        node2 = int(node2)
        weight = int(weight)
        if node1 in AdjacencyDict.keys():
            AdjacencyDict[node1].append((node2, weight))
        else:
            AdjacencyDict[node1] = [(node2,weight)]
            
```

```python
distanceMatrix = calculateDistMatrix(n, AdjacencyDict)
```

```python
for d in distanceMatrix:
    print(' '.join([str(i) for i in d]))
```
