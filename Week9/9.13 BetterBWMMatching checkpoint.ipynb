```python
#####Less memory
def OrderText(Text,k=5):
    n = len(Text)
    TextCount = dict()
    LastColumn= []
    ###Initial Count
    Count = [ {char:0 for char in set(Text)}  for i in range (n//k+1+1) ]    ####A list of dictionary  The first +1 for 0 row, the second +1 is useless for the end
    for i in range(n):
        base = Text[i]
        if(i%k==0):
            for char in set(Text):
                Count[i//k+1][char] = Count[i//k][char]
        Count[i//k+1][base] +=1

        if base in TextCount.keys():
            TextCount[base] += 1
            LastColumn.append((base,TextCount[base]))
        else:
            TextCount[base]  = 0
            LastColumn.append((base,TextCount[base]))
    return LastColumn,TextCount,Count
```

```python
####Update to FisrtOccurence
def Sorted(TextCount):
    FirstOccurence = {}
    i = 0
    for key in sorted(TextCount.keys()):
        val  = TextCount[key]+1
        FirstOccurence[key] = i
        i += val
    return FirstOccurence
```

```python
def Construction(Text):
    n = len(Text)
    Dict1 = dict()
    for i in range(0,n):
        Dict1[Text[i:]] = (i,Text[i-1])
     
    NewText = ''
    SuffixArray = [i for i in range(n)]
    i=0
    for key in sorted(Dict1.keys()):
        index,char = Dict1[key]
        NewText += char
        SuffixArray[i] = index
        i += 1
        
    
    return SuffixArray,NewText
        
```

```python
def BWMatching(SuffixArray,SufText,LastColumn,Pattern,FirstOccurence,Count,k=5):
    top = 0
    bottom = len(LastColumn)-1
    while(top<=bottom):
        if Pattern:
            symbol = Pattern[-1]
            Pattern = Pattern[0:len(Pattern)-1]
            
            ####Update top and bottom
            count = FirstOccurence[symbol] +Count[top//k][symbol]
            ######Modified non exists top in count table
            for i in range(top-top%k,top):
                if (SufText[i] == symbol):
                    count+= 1
            top = count
            
           # if (bottom == len(LastColumn)-1):
             #   if bottom%k==0:
               #     bottom = FirstOccurence[symbol] +Count[(bottom)//k][symbol]-1
                #else:
                   # bottom = FirstOccurence[symbol] +Count[(bottom)//k+1][symbol]-1
                    
            ######Modified non exists top in count table
            #else:
            count =  FirstOccurence[symbol] +Count[(bottom)//k][symbol]
            ######Modified non exists top in count table
            for i in range(bottom-bottom%k ,bottom+1):
                if (SufText[i] == symbol):
                    count+=1
            bottom = count-1
        else:
            #return bottom+1-top
            position = set()
            for i in range(top,bottom+1):
                position.add(SuffixArray[i])
            return position
                
                
                  
    ####No Match
    return {}
```

```python
Text='AATCGGGTTCAATCGGGGT'
Text += '$'
Patterns = ['ATCG' , 'GGGT']
SuffixArray,Text = Construction(Text)

#OriginText='CATTCGACCGCACGGCCGTCCAGAG$'
#Text='GGCGCCGC$TAGTCACACACGCCGTA'
#Patterns = 'ACC CCG CAG'
#Patterns = 'CCG'
#Patterns = Patterns.split(' ')
k=5
LastColumn,TextCount,Count = OrderText(Text,k)
FirstOccurence = Sorted(TextCount)
for Pattern in Patterns:
        Positions = BWMatching(SuffixArray,Text,LastColumn,Pattern,FirstOccurence,Count,k)
        for pos in Positions:
            print(pos,end=' ')
```

```python
def Main(Text,Patterns,k):
    SuffixArray,Text = Construction(Text)
    LastColumn,TextCount,Count = OrderText(Text,k)
    FirstOccurence = Sorted(TextCount)
    for Pattern in Patterns:
        Positions = BWMatching(SuffixArray,Text,LastColumn,Pattern,FirstOccurence,Count,k)
        for pos in Positions:
            print(pos,end=' ')
```
