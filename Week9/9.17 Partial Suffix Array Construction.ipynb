```python
def Construction(Text,k):
    n = len(Text)
    SuffixArray = dict()
    for i in range(0,n):
        SuffixArray[Text[i:]] = Text[i-1]
    return SuffixArray
        
```

```python
def main(Text,k):
    SuffixArray  = Construction(Text,k)
    i=0
    NewText = ''
    for key in sorted(SuffixArray.keys()):
        NewText += SuffixArray[key]
        #if(SuffixArray[key]%k==0):
          #  print(i,SuffixArray[key],sep=',')
        #i+=1
    return NewText
```
