```python
def OrderText(Text):
    n = len(Text)
    TextCount = dict()
    LastColumn= []
    ###Initial Count
    Count = [  {char:0 for char in set(Text)}  for i in range (n+1)]    ####A list of dictionary
    for i in range(n):
        base = Text[i]
        for char in set(Text):
            Count[i+1][char] = Count[i][char]
        Count[i+1][base] +=1
        if base in TextCount.keys():
            TextCount[base] += 1
            LastColumn.append((base,TextCount[base]))
           
            
        else:
            TextCount[base]  = 0
            LastColumn.append((base,TextCount[base]))
    return LastColumn,TextCount,Count
```

```python
####Update to FisrtOccurence
def Sorted(TextCount):
    FirstOccurence = {}
    i = 0
    for key in sorted(TextCount.keys()):
        val  = TextCount[key]+1
        FirstOccurence[key] = i
        i += val
    return FirstOccurence
```

```python
def BWMatching(LastColumn,Pattern,FirstOccurence,Count):
    top = 0
    bottom = len(LastColumn)-1
    while(top<=bottom):
        if Pattern:
            symbol = Pattern[-1]
            Pattern = Pattern[0:len(Pattern)-1]
            
            ####Update top and bottom
            top = FirstOccurence[symbol] + Count[top][symbol]
            bottom =FirstOccurence[symbol] + Count[bottom+1][symbol]-1
        else:
            return bottom+1-top
                  
    ####No Match
    return 0
```

```python
Text='TCCTCTATGAGATCCTATTCTATGAAACCTTCA$GACCAAAATTCTCCGGC'
LastColumn,TextCount,Count = OrderText(Text)
FirstOccurence= Sorted(TextCount)
```

```python
def Main(Text,Patterns):
    LastColumn,TextCount,Count = OrderText(Text)
    FirstOccurence = Sorted(TextCount)
    for Pattern in Patterns:
        match_num = BWMatching(LastColumn,Pattern,FirstOccurence,Count)
        print(match_num,end=' ')
```

```python
Text='GGCGCCGC$TAGTCACACACGCCGTA'
Patterns = 'ACC CCG CAG'
Patterns = Patterns.split(' ')
Main(Text,Patterns)
```

```python
Text=''
Patterns= []
file = open('dataset_369397_7.txt', 'r') 
for i, line in enumerate(file):
    line=line.strip()
    if i==0:
        Text = line
    else:
        Patterns =line.split(' ')
```
